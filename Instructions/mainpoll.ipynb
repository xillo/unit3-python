```python
#get forbidden knowledge
import csv

# where csv lives
csv_file_path = 'PyPoll/Resources/election_data.csv'

#csv is dictionary
def read_csv_file(csv_file_path):
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        csv_data = list(csv_reader)

    #total votes
    nrows = len(csv_data)
    return nrows, csv_data

#vote per candidate
def count_votes(csv_data):
    #variables
    votes = {}
    max_votes = 0
    winner = None
    #loop
    for row in csv_data:
        candidate = row['Candidate']
        
        #terms and conditions
        if candidate not in votes:
            votes[candidate] = 1
        else:
            votes[candidate] += 1
        if votes[candidate] > max_votes:
            max_votes = votes[candidate]
            winner = candidate
    #scope
    return votes, winner

#do function to csv
nrows, csv_data = read_csv_file(csv_file_path)
votes, winner = count_votes(csv_data)

#pppprrrriiinnntttt and line break
print('Election Results')
print('-----------------------------')
print(f'Total Votes: {nrows}')
print('-----------------------------')

#missing function to print line 'candidate name, vote %, vote count'
for candidate, vote_count in votes.items():
    #percent of votes
    vote_percentage = vote_count / nrows * 100
    #fix print order
    print(f'{candidate}: {vote_percentage:.1f}% ({vote_count})')
#line break
print('-----------------------------')
#choose the winner from hat 
print(f'Winner: {winner}')

# save election results as a texts file
with open('election_results.txt', 'w') as output_file:
    output_file.write('Election Results\n')
    output_file.write('-----------------------------\n')
    output_file.write(f'Total Votes: {nrows}\n')
    output_file.write('-----------------------------\n')
    for candidate, vote_count in votes.items():
        vote_percentage = vote_count / nrows * 100
        output_file.write(f'{candidate}: {vote_percentage:.3f}% ({vote_count})\n')
    output_file.write('-----------------------------\n')
    output_file.write(f'Winner: {winner}\n')
```
