```python
#python get forbidden knowledge
import csv

# save where csv lives
csv_file_path = 'Pybank/Resources/budget_data.csv'

# save csv as a library
def read_csv_file(csv_file_path):
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        csv_data = list(csv_reader)

    # Count the rows
    nrows = len(csv_data)

    # Sum of financial activity
    totalsum = sum(int(row['Profit/Losses']) for row in csv_data)

    # average change
    values = []
    for i in range(1, len(csv_data)):
        current_value = csv_data[i]['Profit/Losses']
        previous_value = csv_data[i-1]['Profit/Losses']
        if current_value and previous_value:
            values.append(float(current_value) - float(previous_value))

    avg_change = sum(values) / len(values)

    return (nrows, totalsum, avg_change, csv_data)

#largest change
def find_max_increase(csv_data):
    max_increase = 0
    max_month = ''
    for i in range(1, len(csv_data)):
        current_value = float(csv_data[i]['Profit/Losses'])
        previous_value = float(csv_data[i-1]['Profit/Losses'])
        increase = current_value - previous_value
        if increase > max_increase:
            max_increase = increase
            max_month = csv_data[i]['Date']
    return (max_month, max_increase)

#smolest change
def find_max_decrease(csv_data):
    max_decrease = 0
    max_month = ''
    for i in range(1, len(csv_data)):
        current_value = float(csv_data[i]['Profit/Losses'])
        previous_value = float(csv_data[i-1]['Profit/Losses'])
        decrease = current_value - previous_value
        if decrease < max_decrease:
            max_decrease = decrease
            max_month = csv_data[i]['Date']
    return (max_month, max_decrease)


# do function and remember CSV data
nrows, totalsum, avg_change, csv_data = read_csv_file(csv_file_path)

# find month with max increase in CSV
max_month, max_increase = find_max_increase(csv_data)

# find month with max decrease in CSV
max_month_dec, max_decrease = find_max_decrease(csv_data)

# print the funky results
print('Financial Analysis')
print('-----------------------------')
print(f'Total Months: {nrows}')
print(f'Total: ${totalsum}')
print(f'Average Change: ${avg_change:.2f}')
print(f'Greatest Increase in Profits: {max_month} (${max_increase:.0f})')
print(f'Greatest Decrease in Profits: {max_month_dec} (${max_decrease:.0f})')

# save Financial Analysis to text file
output_file_path = 'Pybank/Analysis/Financial Analysis.txt'
with open(output_file_path, 'w') as output_file:
    output_file.write('Financial Analysis\n')
    output_file.write('-----------------------------\n')
    output_file.write(f'Total Months: {nrows}\n')
    output_file.write(f'Total: ${totalsum}\n')
    output_file.write(f'Average Change: ${avg_change:.2f}\n')
    output_file.write(f'Greatest Increase in Profits: {max_month} (${max_increase:.0f})\n')
    output_file.write(f'Greatest Decrease in Profits: {max_month_dec} (${max_decrease:.0f})\n')
#tell user about the save
print(f'Results written to {output_file_path}')
```
